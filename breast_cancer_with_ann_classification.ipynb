```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df.head()
```

```python
df['label'] = breast_cancer.target
df.head()
```

```python
df['label'].value_counts()
```

```python
df.shape
```

```python
df.info()
```

* Data is **Imbalenced**

**1-->Benign Cancre**

**0-->Malignant Cancre**

### **Seprating the features and target**

```python
x = df.drop(columns='label', axis=1)
y = df['label']
```

### **Spliting the data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
y_train.shape
```

### **Standrize the data**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

### **Bulding the Neural Network**

```python
import tensorflow as tf
from tensorflow import keras
```

```python
# Setting up layers for NN
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)), # It use to to convert multi-dimensional input data into a one-dimensional vector.
                          keras.layers.Dense(20,activation = 'relu'),
                          keras.layers.Dense(30,activation = 'relu'), # Dense layer is a fully connected neural network layer where each neuron is connected to all neurons in the previous layer
                          keras.layers.Dense(20,activation = 'relu'),

                          keras.layers.Dense(2,activation = 'sigmoid')

                          ])
```

```python
model.summary()
```

image.png

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, validation_split=0.1,epochs=10)
```

### **Accuracy of the model on test data**

```python
train_pred = model.predict(x_train)
train_pred[0]
```

### **convert prediction in 0 and 1**

```python
train_pred = [np.argmax(i) for i in  train_pred]
print(train_pred)
```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_score, recall_score
```

```python
test_pred = model.predict(x_test)
test_pred = [np.argmax(i) for i in  test_pred]
print(test_pred)
```

```python
print('Training model accuracy :- ', accuracy_score(y_train, train_pred))
print('training model Confusion Matrix :- \n',confusion_matrix(y_train, train_pred))
print('training model precision_score Report :- ',precision_score(y_train, train_pred))
print('training model recall_score Report :- ',recall_score(y_train, train_pred))
print(' ')
print('*'*80)
print(' ')
print('Testing model accuracy :- ', accuracy_score(y_test, test_pred))
print('Testing model Confusion Matrix :- \n',confusion_matrix(y_test, test_pred))
print('Testing model precision_score Report :- ',precision_score(y_test, test_pred))
print('Testing model recall_score Report :- ',recall_score(y_test, test_pred))
```
