# Spam Mail Prediction

**Import Dependancy**

```python
# Import Dependancy
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('/content/Copy of mail_data.csv')
```

```python
df.head()
```

```python
df.shape
```

### Dealing with null values

**Replacing null values with null string**

```python
df.isnull().sum()
```

### Lable Encoding

*   ham	converted as 1
*   spam converted as 0



```python
# Lable Encoding
df.loc[df['Category'] == 'spam', 'Category',] = 0
df.loc[df['Category'] == 'ham', 'Category',] = 1
df.head()
```

### Train Test Split

**Sepret data as x and y**

```python
x = df['Message']
y = df['Category']
```

**Spliting the data as train data and test data**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

### Feature Extraction

**convert text data into numerical data**

```python
x_train.head()
```

```python
x_test.head()
```

```python
# convert text data into num data
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=True)

# Fit the data
x_train_features = feature_extraction.fit_transform(x_train)
x_test_features = feature_extraction.transform(x_test)
```

```python
print(x_train_features[:5])
```

```python
y_train.info()
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
y_train.info()
```

### Create model

**Logistic regression**

```python
model = LogisticRegression()
model.fit(x_train_features, y_train)
```

### Eveluating the training model

**Prediction score on training data**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score, precision_score, recall_score
```

```python
train_pred = model.predict(x_train_features)
test_pred = model.predict(x_test_features)
print("Accuracy score on training data : ", accuracy_score(y_train, train_pred))
print("Precision Score on training data :", precision_score(y_train, train_pred))
print("Recall Score on training data :", recall_score(y_train, train_pred))
print("Confusion Matrix Score on training data :\n", confusion_matrix(y_train, train_pred))
print(' ')
print('*'*80)
print(' ')
print("Accuracy score on test data : ", accuracy_score(y_test, test_pred))
print("Precision Score on test data :", precision_score(y_test, test_pred))
print("Recall Score on test data :", recall_score(y_test, test_pred))
print("Confusion Matrix Score on test data :\n", confusion_matrix(y_test, test_pred))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(x_train_features, y_train)
```

### Eveluating the training model

**Prediction score on training data**

```python
train_pred = model.predict(x_train_features)
test_pred = model.predict(x_test_features)
print("Accuracy score on training data : ", accuracy_score(y_train, train_pred))
print("Precision Score on training data :", precision_score(y_train, train_pred))
print("Recall Score on training data :", recall_score(y_train, train_pred))
print("Confusion Matrix Score on training data :\n", confusion_matrix(y_train, train_pred))
print(' ')
print('*'*80)
print(' ')
print("Accuracy score on test data : ", accuracy_score(y_test, test_pred))
print("Precision Score on test data :", precision_score(y_test, test_pred))
print("Recall Score on test data :", recall_score(y_test, test_pred))
print("Confusion Matrix Score on test data :\n", confusion_matrix(y_test, test_pred))
```

### Building a Predictive System

```python
input_data = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]

input_data_feature = feature_extraction.transform(input_data)

prediction = model.predict(input_data_feature)
print(prediction)

if prediction[0] == 1:
  print('Ham mail')
else:
  print('Spam mail')
```
